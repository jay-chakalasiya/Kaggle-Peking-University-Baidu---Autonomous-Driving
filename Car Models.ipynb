```python
import os
import json
import numpy as np
PATH = './pku-autonomous-driving/car_models_json/'
import pandas as pd
```

```python
out=[]
for file in os.listdir(PATH):
    with open(PATH+file) as json_file:
        data = np.array(json.load(json_file)['vertices'])
        out.append([file[:-5],np.max(data[:,0]), np.min(data[:,0]), np.max(data[:,1]), np.min(data[:,1]), np.max(data[:,2]), np.min(data[:,2])])        
```

```python

df = pd.DataFrame(out, columns = ['Model','X-max', 'X-min', 'Y-max', 'Y-min', 'Z-max', 'Z-min'])
df.to_csv('Car_dimentions')
```
