```python
def rotation_matrix(roll, pitch, yaw):
    cu, cv, cw = np.cos(float(roll)), np.cos(float(pitch)), np.cos(float(yaw))
    su, sv, sw = np.sin(float(roll)), np.sin(float(pitch)), np.sin(float(yaw))
    return np.array([[cv*cw, su*sv*cw-cu*sw, su*sw+cu*sv*cw],
                     [cv*sw, cu*cw+su*sv*sw, cu*sv*sw-su*cw],
                     [-1*sv, su*cv, cu*cv]])
def traslation_rotation_matrix(rotation_matrix, traslation_matrix):
    out = np.zeros((4,4))
    out[:3, :3] = rotation_matrix
    out[:3, 3:] = np.array([traslation_matrix]).T
    out[3,3] = 1
    return out
                      
```

```python
DATA_PATH = './pku-autonomous-driving/'
train = pd.read_csv(DATA_PATH + 'train.csv')

def str2coords(s, names=['id', 'yaw', 'pitch', 'roll', 'x', 'y', 'z']):
    coords = []
    extra_coords = []
    
    for l in np.array(s.split()).reshape([-1, 7]):
        coords.append(dict(zip(names, l.astype('float'))))
        if 'id' in coords[-1]:
            coords[-1]['id'] = int(coords[-1]['id'])
        car_name = Id_to_Car[int(l[0])]
        
        Transformation_matrix = rotation_matrix(l[3], float(l[2]), l[1])
        
        X_max = np.dot(Transformation_matrix, [float(car_extra_point_data.loc[car_extra_point_data['Model'] == car_name]['X-max']), 0, 0])
        X_min = np.dot(Transformation_matrix, [float(car_extra_point_data.loc[car_extra_point_data['Model'] == car_name]['X-min']), 0, 0])
        Y_max = np.dot(Transformation_matrix, [0, float(car_extra_point_data.loc[car_extra_point_data['Model'] == car_name]['Y-max']), 0])
        Y_min = np.dot(Transformation_matrix, [0, float(car_extra_point_data.loc[car_extra_point_data['Model'] == car_name]['Y-min']), 0])
        Z_max = np.dot(Transformation_matrix, [0, 0, float(car_extra_point_data.loc[car_extra_point_data['Model'] == car_name]['Z-max'])])
        Z_min = np.dot(Transformation_matrix, [0, 0, float(car_extra_point_data.loc[car_extra_point_data['Model'] == car_name]['Z-min'])])
        
        
        extra_coords.append(dict(zip(['X-maxX', 'X-maxY', 'X-maxZ',
                                      'X-minX', 'X-minY', 'X-minZ',
                                      'Y-maxX', 'Y-maxY', 'Y-maxZ',
                                      'Y-minX', 'Y-minY', 'Y-minZ',
                                      'Z-maxX', 'Z-maxY', 'Z-maxZ',
                                      'Z-minX', 'Z-minY', 'Z-minZ'], 
                                     [X_max[0]+float(l[4]), X_max[1]+float(l[5]), X_max[2]+float(l[6]),
                                      X_min[0]+float(l[4]), X_min[1]+float(l[5]), X_min[2]+float(l[6]),
                                      Y_max[0]+float(l[4]), Y_max[1]+float(l[5]), Y_max[2]+float(l[6]),
                                      Y_min[0]+float(l[4]), Y_min[1]+float(l[5]), Y_min[2]+float(l[6]),
                                      Z_max[0]+float(l[4]), Z_max[1]+float(l[5]), Z_max[2]+float(l[6]),
                                      Z_min[0]+float(l[4]), Z_min[1]+float(l[5]), Z_min[2]+float(l[6])])))        
    return coords, extra_coords
```

```python
def imread(path, fast_mode=False):
    img = cv2.imread(path)
    if not fast_mode and img is not None and len(img.shape) == 3:
        img = np.array(img[:, :, ::-1])
    return img
```

```python
def get_img_coords(s):
    '''
    Input is a PredictionString (e.g. from train dataframe)
    Output is two arrays:
        xs: x coordinates in the image (row)
        ys: y coordinates in the image (column)
    '''
    coords, _ = str2coords(s)
    xs = [c['x'] for c in coords]
    ys = [c['y'] for c in coords]
    zs = [c['z'] for c in coords]
    P = np.array(list(zip(xs, ys, zs))).T
    img_p = np.dot(CAMERA_MATRIX, P).T
    img_p[:, 0] /= img_p[:, 2]
    img_p[:, 1] /= img_p[:, 2]
    img_xs = img_p[:, 0]
    img_ys = img_p[:, 1]
    img_zs = img_p[:, 2] # z = Distance from the camera
    return img_xs, img_ys

plt.figure(figsize=(14,14))
plt.imshow(imread(DATA_PATH + 'train_images/' + train['ImageId'][2217] + '.jpg'))
plt.scatter(*get_img_coords(train['PredictionString'][2217]), color='red', s=100);
```

```python
def get_extra_img_coords(s, dim_string='x-mx'):
    '''
    Input is a PredictionString (e.g. from train dataframe)
    Output is two arrays:
        xs: x coordinates in the image (row)
        ys: y coordinates in the image (column)
    '''
    ID = 'X-max'
    
    if dim_string=='x-mx':
        ID='X-max'
    elif dim_string=='x-mn':
        ID='X-min'
    elif dim_string=='y-mx':
        ID='Y-max'
    elif dim_string=='y-mn':
        ID='Y-min'
    elif dim_string=='z-mx':
        ID='Z-max'
    elif dim_string=='z-mn':
        ID='Z-min'
    
    _, coords = str2coords(s)
    xs = [c[ID+'X'] for c in coords]
    ys = [c[ID+'Y'] for c in coords]
    zs = [c[ID+'Z'] for c in coords]
    P = np.array(list(zip(xs, ys, zs))).T
    img_p = np.dot(CAMERA_MATRIX, P).T
    img_p[:, 0] /= img_p[:, 2]
    img_p[:, 1] /= img_p[:, 2]
    img_xs = img_p[:, 0]
    img_ys = img_p[:, 1]
    img_zs = img_p[:, 2] # z = Distance from the camera
    return img_xs, img_ys

plt.figure(figsize=(14,14))
img = 2211
plt.imshow(imread(DATA_PATH + 'train_images/' + train['ImageId'][img] + '.jpg'))
plt.scatter(*get_img_coords(train['PredictionString'][img]), color='red', s=50);
plt.scatter(*get_extra_img_coords(train['PredictionString'][img], dim_string='x-mx'), color='green', s=50);
plt.scatter(*get_extra_img_coords(train['PredictionString'][img], dim_string='x-mn'), color='green', s=50);
plt.scatter(*get_extra_img_coords(train['PredictionString'][img], dim_string='y-mx'), color='yellow', s=50);
plt.scatter(*get_extra_img_coords(train['PredictionString'][img], dim_string='y-mn'), color='yellow', s=50);
plt.scatter(*get_extra_img_coords(train['PredictionString'][img], dim_string='z-mx'), color='pink', s=50);
plt.scatter(*get_extra_img_coords(train['PredictionString'][img], dim_string='z-mn'), color='pink', s=50);
```
